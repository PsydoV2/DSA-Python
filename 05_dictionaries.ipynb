# Dictionaries (dict)

- Key/Value pairs
- Associative array, like Java HashMap
- Dicts are Unordered

**constructors**

```python
x = {"pork": 25.3, "beef": 33.8, "chicken": 22.7}
print(x)
```

```python
x = dict([("pork", 25.3), ("beef", 33.8), ("chicken", 22.7)])
print(x)
```

```python
x = dict(pork = 25.3, beef = 33.8, chicken = 22.7)
print(x)
```

**dict operations**

```python
x["shrimp"] = 38.2
print(x)
```

```python
del(x["shrimp"])
print(x)
```

```python
print(len(x))
```

```python
x.clear()
print(x)
```

```python
del(x)
```

**accessing keys and values in a dict**

```python
y = {"pork": 25.3, "beef": 33.8, "chicken": 22.7}
print(y.keys())
print(y.values())
print(y.items())
```

```python
print("beef" in y)
```

```python
print("clams" in y.values())
```

**iterating a dict** - note, items are in random order

```python
for key in y:
    print(key, y[key])
```

```python
for k, v in y.items():
    print(k ,v)
```
