# Sets

- Store non-duplicate items
- Very fast access vs Lists
- Math Set ops (union, intersect)
- Sets are **Unordered**

**constructors**

```python
x = {3, 5, 3, 5}
print(x)
```

```python
y = set()
print(y)
```

```python
list = [2, 3, 4]
z = set(list)
print(z)
```

**set operations**

```python
x = {3, 8, 5}
print(x)

x.add(7)
print(x)
```

```python
x = {3, 8, 5}
print(x)

x.remove(3)
print(x)
```

```python
x = {3, 8, 5}
print(len(x))
```

```python
x = {3, 8, 5}
print(5 in x)
```

```python
x = {3, 8, 5}
print(x.pop(), x)   # random item gets removed because sets are not ordered so we don't know which item is last!
```

```python
x = {3, 8, 5}
x.clear()
print(x)
```

**Mathematical set operations**

```python
s1 = {1, 2, 3}
s2 = {3, 4, 5}

print(s1 & s2)      # AND   returning items which are in both
print(s1 | s2)      # OR    returning all items without duplicates because sets don't have duplicates
print(s1 ^ s2)      # XOR   returning all items which are in one but not in both
print(s1 - s2)      # in set1 but not in set 2
print(s2 - s1)      # in set2 but not in set 1
print(s1 <= s2)     # subset (set2 contains set1?)
print(s1 >= s2)     # superset (set1 contains set2?)
```
