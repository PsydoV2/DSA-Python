# Sequences: String, List, Tuple

### Indexing

```python
# string

x = 'frog'
print(x[3])
```

```python
# list

x = ["pig", "cow", "horse"]
print(x[1])
```

```python
# tuple

x = ("Kevin", "Niklas", "Jenny", "Craig")
print(x[0])
```

### Slicing

```python
x = "computer"

print(x[1:4])
print(x[1:6:2])
print(x[0::2])
print(x[3:])
print(x[:5])
print(x[-1])
print(x[-3:])
print(x[:-2])
```

### adding & concatentaning

```python
# string

x = "horse" + "shoe"
print(x)
```

```python
# list

y = ["pig", "cow"] + ["horse"]
print(y)
```

```python
#tuple

z = ("Kevin", "Niklas", "Jenny") + ("Craig",) # Komma am Ende das es sonst als String gesehen wird und nicht als tuple und tuple + string geht nicht!
print(z)
```

### multiplying

```python
# string

x = "bug" * 3
print(x)
```

```python
# list
y = [8, 5] * 3
print(y)
```

```python
# tuple
z = (2, 4) * 3
print(z)
```

### checking membership

```python
# string

x = "bug"
print("u" in x)
```

```python
# list

y = ["pig", "cow", "horse"]
print("cow" not in y)
```

```python
# tuple

z = ("Kevin", "Niklas", "Jenny", "Craig")
print("Niklas" in z)
```

### iterating

```python
# item

x = [7, 8, 3]
for item in x:
    print(item)
```

```python
# index & item

y = [7, 8, 3]
for index, item in enumerate(y):
    print(index, item)
```
