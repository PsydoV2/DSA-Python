# Sequences: String, List, Tuple

**Indexing**

```python
# string

x = 'frog'
print(x[3])
```

```python
# list

x = ["pig", "cow", "horse"]
print(x[1])
```

```python
# tuple

x = ("Kevin", "Niklas", "Jenny", "Craig")
print(x[0])
```

**Slicing**

```python
x = "computer"

print(x[1:4])
print(x[1:6:2])
print(x[0::2])
print(x[3:])
print(x[:5])
print(x[-1])
print(x[-3:])
print(x[:-2])
```

**adding & concatentaning**

```python
# string

x = "horse" + "shoe"
print(x)
```

```python
# list

y = ["pig", "cow"] + ["horse"]
print(y)
```

```python
#tuple

z = ("Kevin", "Niklas", "Jenny") + ("Craig",) # Komma am Ende das es sonst als String gesehen wird und nicht als tuple und tuple + string geht nicht!
print(z)
```

**multiplying**

```python
# string

x = "bug" * 3
print(x)
```

```python
# list
y = [8, 5] * 3
print(y)
```

```python
# tuple
z = (2, 4) * 3
print(z)
```

**checking membership**

```python
# string

x = "bug"
print("u" in x)
```

```python
# list

y = ["pig", "cow", "horse"]
print("cow" not in y)
```

```python
# tuple

z = ("Kevin", "Niklas", "Jenny", "Craig")
print("Niklas" in z)
```

**iterating**

```python
# item

x = [7, 8, 3]
for item in x:
    print(item)
```

```python
# index & item

y = [7, 8, 3]
for index, item in enumerate(y):
    print(index, item)
```

**number of items**

```python
# string

x = "bug"
print(len(x))
```

```python
# list

y = ["pig", "cow", "horse"]
print(len(y))
```

```python
# tuple

z = ("Kevin", "Niklas", "Jenny", "Craig")
print(len(z))
```

**minimum**

```python
# string

x = 'bug'
print(min(x))
```

```python

# list

y = ['pig', 'cow', 'horse' ]
print(min(y))
```

```python

# tuple

z = ('Kevin', 'Niklas', 'Jenny', 'Craig' )
print(min(z))
```

**maximum**

```python
# string

x = 'bug'
print(max(x))
```

```python
# list

y = ['pig', 'cow', 'horse' ]
print(max(y))
```

```python
# tuple

z = ('Kevin', 'Niklas', 'Jenny' , 'Craig' )
print(max(z))
```

**sum**

```python
# string -> error
x = [5, 7, 'bug' ]
print (sum(x))      # generates an error
```

```python
# list

y = [2, 5, 8, 12]
print(sum(y))
print(sum(y[-2:]))
```

```python
# tuple

z = (50, 4, 7, 19)
print(sum(z))
```

**sorting**

```python
# string

x = 'bug'
print(sorted(x))
```

```python
# list

y = ['pig', 'cow', 'horse' ]
print(sorted(y))
```

```python
# tuple

z = ('Kevin' , 'Niklas', 'Jenny' , 'Craig' )
print(sorted(z))
```

**sorting by second letter**

```python
z = ("Kevin", "Niklas", "Jenny", "Craig")
print(sorted(z, key=lambda k: k[1:]))
```

**count(item)**

```python
# string

x = 'hippo'
print(x.count('p'))
```

```python
# list

y = ['pig', 'cow', 'horse', 'cow' ]
print(y.count('cow'))
```

```python
# tuple

z = ('Kevin' , 'Niklas', 'Jenny' , 'Craig' )
print(z.count('Kevin'))
```

**index(item)**

```python
# string

x = 'hippo'
print(x.index('p'))
```

```python
# list

y = ['pig', 'cow', 'horse', 'cow' ]
print(y.index('cow'))
```

```python
# tuple

z = ('Kevin', 'Niklas', 'Jenny' , 'Craig' )
print(z.index('Jenny'))
```

**unpacking**

```python
x = ["pig", "cow", "horse"]
a, b, c = x
print(a, b, c)
```
