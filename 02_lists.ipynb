# Lists

- General purpose
- Most widely used data structure
- Grow and shrink size as needed
- Sequence type
- Sortable

**constructors**

```python
x = list()
y = ["a", 25, "dog", 8.43]
tuplel = (10, 20)
z = list(tuple)
```

```python
# list comprehension

a = [m for m in range(8)]
print(a)

b = [i**2 for i in range(10) if i>4] # i^2 for i greater 4 is written in the list (5^2, 6^2, ...)
print(b)
```

**delete**

```python
x = [5, 3, 8, 6]
del(x[1])
print(x)
del(x)      # list x no longer exists
```

**append**

```python
x = [5, 3, 8, 6]
x.append(7)
print(x)
```

**extend**

```python
x = [5, 3, 8, 6]
y = [12, 13]
x.extend(y)
print(x)
```

**insert**

```python
x = [5, 3, 8, 6]
print(x)

x.insert(1, 7)
print(x)

x.insert(1, ["a", "m"])
print(x)
```

**pop**

```python
x = [5, 3, 8, 6]
print(x)

x.pop()
print(x)

print(x.pop())
```

**remove**

```python
x = [5, 3, 8, 6, 3]
x.remove(3)
print(x)
```

**revers**

```python
x = [5, 3, 8, 6]
x.reverse()
print(x)
```

**sort**

```python
# sorted(x) returns a new list
# x.sort() sorts the order in place without creating a new list

x = [5, 3, 8, 6]
x.sort()
print(x)
```

**reverse sort**

```python
x = [5, 3, 8, 6]
x.sort(reverse=True)
print(x)
```
