# Tuples

- Immutable (can't add/change)
- Useful for fixed data
- Faster than Lists
- Sequence type

**constructors**

```python
x = ()
x = (1, 2, 3)
x = 1, 2, 3
x = 2,  # without the comma python would interpret it as a integer!
print(x, type(x))
```

```python
list = [2, 4, 6]
x = tuple(list)
print(x, type(x))
```

**tuples are immutable**, but member objects may be mutable 

```python
x = (1, 2, 3)
del(x[1])         # fails
[1] = 8          # fails
print(x)
```

```python
y = ([1, 2], 3)     # tuple where first item is a list
print(y)

del(y[0][1])        # deleting the index 1 of the member at index 0 in the tuple (the 2)
print(y)

y += (4,)           # concatenating two tuples
print(y)
```
